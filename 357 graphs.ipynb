```python
counts_mean=[]
voltage=[700.2,750.7,800,850.4,899.2,951.7]


counts=[[6,5,6],
        [6,4,1],
        [9,8,4],
        [12,8,7],
        [5,8,8],
        [34,32,15]]


for i in range(0,len(counts)):
    print(np.mean(counts[i]))
    counts_mean.append(np.mean(counts[i]))
    
plt.plot(voltage,counts_mean, "x")
plt.savefig('Radioactivity_V_vs._Count.pdf', format='pdf', bbox_inches='tight', transparent=True)
plt.show()
```

```python
count=[11,5,7,4,5,3,6,5,5,8,5,6,3,5,6,8,6,6,9,7,8,10,6,7,10,6,4,2,11]
time=list(range(30, (len(count)+1)*30, 30))

plt.plot(time, count, "x")
plt.xlabel("seconds")
plt.ylabel('counts in that interval')
plt.figtext(0.2, 0.8, "Average count:%s" %('%0.8f'%np.mean(count)))
plt.savefig('Radioactivity_background_count.pdf', format='pdf', bbox_inches='tight', transparent=True)

print("Average count:", np.mean(count))
```

```python
count_max=max(count)+2
dist=[]
for i in range(0,count_max):
    dist.append(count.count(i))
    
plt.plot(list(range(0,count_max)), dist, "x")
plt.ylabel("Number of occurrences")
plt.xlabel("Count")
plt.savefig('Radioactivity_background_count_density.pdf', format='pdf', bbox_inches='tight', transparent=True)
```
