```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = np.genfromtxt('expthediff.dat',usecols=0) #Number of neutrons
Z = np.genfromtxt('expthediff.dat',usecols=1) #Number of protons
A = N+Z # Total number of Nucleons
# Difference between the experimental binding energy and the Liquid drop model
y = np.genfromtxt('expthediff.dat',usecols=2)

```

```python
plt.plot(N,y,'o',markersize=3)
plt.show()
```

```python
plt.plot(Z,y,'o',markersize=3)
plt.show()
```

```python
plt.plot(A,y,'o',markersize=3)
plt.show()
```

```python
ext = np.int(np.ceil(max(-np.min(y),np.max(y))))
fig=plt.figure(figsize=(18, 8), dpi= 80, facecolor='w', edgecolor='k')
sc = plt.scatter(N,Z,c=y,cmap='PiYG',vmin=-ext,vmax=ext,s=10)
plt.colorbar(sc)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.transpose(np.stack((N,Z)))

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(X_train) # This sets the shifting and scaling so the training features have mean zero and variance 1

Xs_train = scaler.transform(X_train) # Scaling the training features

Xs_test = scaler.transform(X_test) #Using the SAME scaling on the test features in order to get the correct score
```

```python
from sklearn import linear_model
```

```python
least_squares = linear_model.LinearRegression()
least_squares.fit(Xs_train,y_train)
print(least_squares.coef_)
#print("Training score:", least_squares.score(Xs_train,y_train))
#print("Test score:", least_squares.score(Xs_test,y_test))
pred_train = least_squares.predict(Xs_train)
pred_test = least_squares.predict(Xs_test)
Xs = scaler.transform(X)
pred_all = least_squares.predict(Xs) 
fig=plt.figure(figsize=(18, 8), dpi= 80, facecolor='w', edgecolor='k')
sc = plt.scatter(np.transpose(X)[0],np.transpose(X)[1],c=pred_all,cmap='PiYG',vmin=-ext,vmax=ext,s=10)
plt.colorbar(sc)
plt.show()

plt.plot(N+Z,y,'o',markersize=3,label='Original')
plt.plot(np.transpose(X)[0]+np.transpose(X)[1],pred_all,'o',markersize=3,label='Trained')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ls_score_train = []
ls_score_test = []
degrees = np.arange(1,50)

for deg in degrees:
    poly = PolynomialFeatures(deg)
    X_poly = poly.fit_transform(X) #X is NZ values in array, X_polly is 1, N, Z, N^2, NZ, Z^2 
    Xp_train, Xp_test, yp_train, yp_test = train_test_split(X_poly,y)
    
    poly_scaler = StandardScaler().fit(Xp_train)
    Xps_train = poly_scaler.transform(Xp_train)
    
    Xps_test  = poly_scaler.transform(Xp_test)
    least_squares_p = linear_model.LinearRegression()
    least_squares_p.fit(Xps_train,yp_train)
    ls_score_train.append(least_squares_p.score(Xps_train,yp_train))
    ls_score_test.append(least_squares_p.score(Xps_test,yp_test))
    
plt.plot(degrees,ls_score_train,label='Train')
plt.plot(degrees,ls_score_test,label='Test')
plt.legend()
plt.show()
    
Xps = poly_scaler.transform(X_poly)
pred_all = least_squares_p.predict(Xps)
fig=plt.figure(figsize=(18, 8), dpi= 80, facecolor='w', edgecolor='k')
sc = plt.scatter(np.transpose(X)[0],np.transpose(X)[1],c=pred_all,cmap='PiYG',vmin=-ext,vmax=ext,s=10)
plt.colorbar(sc)
plt.show()

plt.plot(N+Z,y,'o',markersize=3,label='Original')
plt.plot(np.transpose(X)[0]+np.transpose(X)[1],pred_all,'o',markersize=3,label='Trained')
plt.legend()
plt.show()
```

```python
ls_score_train = []
ls_score_test = []
degrees = np.arange(1,50)
for deg in degrees:
    poly = PolynomialFeatures(deg)
    X_poly = poly.fit_transform(X)
    Xp_train, Xp_test, yp_train, yp_test = train_test_split(X_poly,y)
    poly_scaler = StandardScaler().fit(Xp_train)
    Xps_train = poly_scaler.transform(Xp_train)
    Xps_test = poly_scaler.transform(Xp_test)
    least_squares_p = linear_model.Ridge(alpha=0.0001)
    least_squares_p.fit(Xps_train,yp_train)
    ls_score_train.append(least_squares_p.score(Xps_train,yp_train))
    ls_score_test.append(least_squares_p.score(Xps_test,yp_test))


plt.plot(degrees,ls_score_train,label='Train')
plt.plot(degrees,ls_score_test,label='Test')
plt.legend()
plt.show()
    
Xps = poly_scaler.transform(X_poly)
pred_all = least_squares_p.predict(Xps)
fig=plt.figure(figsize=(18, 8), dpi= 80, facecolor='w', edgecolor='k')
sc = plt.scatter(np.transpose(X)[0],np.transpose(X)[1],c=pred_all,cmap='PiYG',vmin=-ext,vmax=ext,s=10)
plt.colorbar(sc)
plt.show()

plt.plot(N+Z,y,'o',markersize=3,label='Original')
plt.plot(np.transpose(X)[0]+np.transpose(X)[1],pred_all,'o',markersize=3,label='Trained')
plt.legend()
plt.show() 
```

```python
alphas = np.logspace(-7, 7, 20)
ls_score_train = []
ls_score_test = []
print(alphas)
degrees = np.arange(1,50)
best_alphas = []
for deg in degrees:
    poly = PolynomialFeatures(deg)
    X_poly = poly.fit_transform(X)
    Xp_train, Xp_test, yp_train, yp_test = train_test_split(X_poly,y)
    poly_scaler = StandardScaler().fit(Xp_train)
    Xps_train = poly_scaler.transform(Xp_train)
    Xps_test = poly_scaler.transform(Xp_test)
    least_squares_p = linear_model.RidgeCV(alphas=alphas)
    least_squares_p.fit(Xps_train,yp_train)
    ls_score_train.append(least_squares_p.score(Xps_train,yp_train))
    ls_score_test.append(least_squares_p.score(Xps_test,yp_test))
    best_alphas.append(least_squares_p.alpha_)

print(best_alphas)
plt.semilogy(degrees,best_alphas)
plt.show()

plt.plot(degrees,ls_score_train,label='Train')
plt.plot(degrees,ls_score_test,label='Test')
plt.legend()
plt.show()
    
Xps = poly_scaler.transform(X_poly)
pred_all = least_squares_p.predict(Xps)
fig=plt.figure(figsize=(18, 8), dpi= 80, facecolor='w', edgecolor='k')
sc = plt.scatter(np.transpose(X)[0],np.transpose(X)[1],c=pred_all,cmap='PiYG',vmin=-ext,vmax=ext,s=10)
plt.colorbar(sc)
plt.show()

plt.plot(N+Z,y,'o',markersize=3,label='Original')
plt.plot(np.transpose(X)[0]+np.transpose(X)[1],pred_all,'o',markersize=3,label='Trained')
plt.legend()
plt.show()    
```
