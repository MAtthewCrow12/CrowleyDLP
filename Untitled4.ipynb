```python
#Importing essentials
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import norm
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import normalize
from sklearn.metrics import mean_squared_error

#Graph Settings
fig_size = plt.rcParams["figure.figsize"]
# Prints: [8.0, 6.0]
#print ("Current size:", fig_size)
# Set figure width to 12 and height to 9
fig_size[0] = 8
fig_size[1] = 12
plt.rcParams["figure.figsize"] = fig_size
```

```python
height=15

x=list(range(0,height+1,1))
y=list(range(0,height+1,1))
x=x+list(range(height,-1,-1))
y=y+list(range(0,height+1,1))
z=[]
xy=[]


#x=x+list(range(15,101,25))
#y=y+list(range(15,101,25))

    

for i in range(0,len(x)):
    xy.append((x[i],y[i]))
    z.append(x[i]+y[i])

print(x)
print(y)
print(z)
print(xy)
```

```python
#y=x+25
#y=x-25
train_test=[]
train_test_target=[]
upper_space=[]
upper_space_target=[]
lower_space=[]
lower_space_target=[]


square=250
strip_width=25

for x in range(0,square):
    for y in range(0,square):
        if y<x+strip_width and y>x-strip_width:
            x_p1=x-125
            y_p1=y-125
            x_p2=x-0
            y_p2=y-200
            x_p3=x-200
            y_p3=y-0
            
            
            train_test.append((x,y,x_p1,y_p1,x_p2,y_p2,x_p3,y_p3))
            train_test_target.append(x+y)
        elif y>=x+strip_width:
            x_p1=x-125
            y_p1=y-125
            x_p2=x-0
            y_p2=y-200
            x_p3=x-200
            y_p3=y-0
            
            
            upper_space.append((x,y,x_p1,y_p1,x_p2,y_p2,x_p3,y_p3))
            upper_space_target.append(x+y)
        else:
            x_p1=x-125
            y_p1=y-125
            x_p2=x-0
            y_p2=y-200
            x_p3=x-200
            y_p3=y-0
            
            lower_space.append((x,y,x_p1,y_p1,x_p2,y_p2,x_p3,y_p3))
            lower_space_target.append(x+y)
```

```python
#plt.plot(np.transpose(train_test)[0],np.transpose(train_test)[1], "x")
#plt.plot(np.transpose(upper_space)[0],np.transpose(upper_space)[1], "+")
#plt.plot(np.transpose(lower_space)[0],np.transpose(lower_space)[1], "+")
strip_err_avg=[]
upper_err_avg=[]
lower_err_avg=[]
strip_sco_avg=[]
upper_sco_avg=[]
lower_sco_avg=[]
    
rf_object=RandomForestRegressor(n_estimators=10000)#, bootstrap=False,)
X_train, X_test, y_train, y_test = train_test_split(train_test, train_test_target, test_size =0.25)#, random_state=42 )
rf_object.fit(X_train,y_train)

strip=rf_object.predict(train_test)
upper=rf_object.predict(upper_space)
lower=rf_object.predict(lower_space)

print("train score:  ", rf_object.score(X_train, y_train))
print("test score:   ", rf_object.score(X_test, y_test), "\n")

print("strip score:  ", rf_object.score(train_test,train_test_target))
print("upper score:  ", rf_object.score(upper_space,upper_space_target))
print("lower score:  ", rf_object.score(lower_space,lower_space_target), "\n")
print("strip error:  ", np.sqrt(mean_squared_error(strip,train_test_target)))
print("upper error:  ", np.sqrt(mean_squared_error(upper,upper_space_target)))
print("lower error:  ", np.sqrt(mean_squared_error(lower,lower_space_target)), "\n")



strip_err_avg.append(rf_object.score(train_test,train_test_target))
upper_err_avg.append(rf_object.score(upper_space,upper_space_target))
lower_err_avg.append(rf_object.score(lower_space,lower_space_target))
strip_sco_avg.append(np.sqrt(mean_squared_error(strip,train_test_target)))
upper_sco_avg.append(np.sqrt(mean_squared_error(upper,upper_space_target)))
lower_sco_avg.append(np.sqrt(mean_squared_error(lower,lower_space_target)))





all_values=train_test+upper_space+lower_space

all_predictions=list(strip)+list(upper)+list(lower)
all_actual=train_test_target+upper_space_target+lower_space_target
```
